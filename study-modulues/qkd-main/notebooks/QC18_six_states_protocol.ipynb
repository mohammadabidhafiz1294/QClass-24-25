# Six-State Protocol:
Six-state protocol distibutes private key through public channel for message encryption through one-time pad. It is like BB84 protocol, which uses single pulses and is based on Heisenberg's uncertainty principle, which makes it impossible for an adversary to measure the incoming photon with certainty and without introducingi an error. This protocol has an advantage over BB84 that the adversary gains less information and it can tolerate more qubit error rate. In this protocol we use 3 basis sets (6 states) while in BB84 we use 2 basis (4 states). In six state protocol we use $X$, $Y$ and $Z$ basis. 
### Z-Basis
The Z basis or the standard basis can be understood as vertically and horizontally polarized photons: 
 $$ \left| h \right\rangle  =  \left| 0 \right\rangle $$ 
 $$ \left| v \right\rangle  =  \left| 1 \right\rangle $$ 
### X-Basis
The X basis or the Hadamard basis can be understood as diagonally polarized light: 
 $$ \left| l \right\rangle  =  \frac{1}{\sqrt{2}}(\left| 0 \right\rangle + \left| 1 \right\rangle) $$ 
 $$ \left| r \right\rangle  =  \frac{1}{\sqrt{2}}(\left| 0 \right\rangle - \left| 1 \right\rangle) $$
### Y-Basis
The Y basis or the circuler basis can be understood as circullarly polarized light:
 $$ \left| u \right\rangle  =  \frac{1}{\sqrt{2}}(\left| 0 \right\rangle + i\left| 1 \right\rangle) $$ 
 $$ \left| w \right\rangle  =  \frac{1}{\sqrt{2}}(\left| 0 \right\rangle - i\left| 1 \right\rangle) $$

### Interconvertion of Basis
The $Z$ basis state can be coverted to $X$ basis by applying Hadamard gate $H$. The $Z$ basis can be coverted to $Y$ basis state by applying the Hadamard gate $H$ and then the phase gate $S$ given as: 

$$
H=
\frac{1}{\sqrt{2}}\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix}
$$

$$
S=
\begin{pmatrix}
1 & 0 \\
0 & i
\end{pmatrix}
$$




## The Six-state Protocol goes as follows:
1. Asja and Balvis agree on storing the bit 0 in $h$, $l$ and $u$ states while storing the 1 bit in $v$, $R$ and $w$ states. 
2. **For Asja:**
 - Asja creates an $n$-bit long random bit string and prepares qubits.
 - Asja encodes qubits in a basis ( $Z$, $X$ or  $Y$ ) chosen randomly:
    -  does nothing if she chose  $Z$-basis,
    -  applies $H$-gate if she chose  $X$-basis.
    -  applies $H$ and then $S$ gate if she chose $Y$ basis
 - Asja saves information about initial bit value and corresponding basis she used to encode that bit.
 - Asja sends qubits to Balvis.

3. **For Balvis:**
 - Balvis, randomly and independently from Asja, chooses a basis in which he will measure each received qubit.
 - Balvis measures qubits in chosen basis ( $Z$, $X$ or $Y$ ) interchanging the basis as Asja does.
 - Balvis saves information about the basis he used and corresponding measurement results.

4. **Basis Reconciliation and Sifting:**
 - Asja and Balvis announce the basis they used for each qubit.
 - They discard the bits where they used different basis. 2/3 of the bits are discarded.
 - They keep the bits where they used the same basis
 - If there is no eavesdropper, their bit string exactly matches.

5. **Error Correction and Privacy Amplification:**
 - If there is an Eavesdropper, she will have 1/3 probability of choosing the same basis as Asja and Balvis, for intercept-resend attack, while in BB84 the probability was 1/2. Her information will be 33% compared to 50% in BB84.
 - The interception will introduce errors. Alice and Bob do the error estimation, then classical error correction and privacy amplification as in BB84 protocol. 

**In this cell we define some important functions and generate basis for Asja and Balvis.**

```python


# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, assemble
from qiskit.quantum_info import Operator
from qiskit.circuit import Parameter
from qiskit.qasm3 import dumps
from qiskit_ibm_runtime import Sampler
from qiskit.compiler import transpile
from qiskit_aer import Aer
from random import randrange
from qiskit.primitives import BackendSampler
import hashlib 



def print_outcomes_in_reserve(counts): # takes a dictionary variable
    for outcome in counts: # for each key-value in dictionary
        reverse_outcome = ''
        for i in outcome: # each string can be considered as a list of characters
            reverse_outcome = i + reverse_outcome # each new symbol comes before the old symbol(s)
    return reverse_outcome

#Source for SendState: awards/teach_me_qiskit_2018/cryptography/Cryptography.ipynb



def SendState(qc1, qc2, qc1_name):
    ''' This function takes the output of a circuit qc1 (made up only of x and 
        h gates and initializes another circuit qc2 with the same state
    ''' 
    
    qs = dumps(qc1).split(sep=';')[4:-1]
    print(qs)

    

    # Process the code to get the instructions 
    for index, instruction in enumerate(qs):
        qs[index] = instruction.lstrip() #replaces the original instruction(string odf quantum instruction) with the striped version

    # Parse the instructions and apply to new circuit
    for instruction in qs:
       
        if instruction[0] == 'x':
            if instruction[5] == '[':
                    old_qr = int(instruction[6:-1])
            else:
                old_qr = int(instruction[5:-1])
            qc2.x(qreg[old_qr])
        elif instruction[0] == 'h':
            if instruction[5] == '[':
                old_qr = int(instruction[6:-1])
            else:
                old_qr = int(instruction[5:-1])
            qc2.h(qreg[old_qr])
        elif instruction[0] == 's':
            if instruction[5] == '[':
                old_qr = int(instruction[6:-1])
            else:
                old_qr = int(instruction[5:-1])
            qc2.s(qreg[old_qr])

        elif instruction[0] == 'm': # exclude measuring:
            pass
        else:
            raise Exception('Unable to parse instruction')

qreg = QuantumRegister(24) # quantum register with 24 qubits
creg = ClassicalRegister(24) # classical register with 24 bits

# Quantum circuit for Asja state
asja = QuantumCircuit(qreg, creg, name='Asja')

send=[] #Initial bit string to send
asja_basis=[] #Initial Asja Basis
balvis_basis=[] #Initial Balvis basis

# Creating randomly chosen basis for Asja (X,Y,Z)
for i in range(24):
    a_b = randrange(3)
    asja_basis.append(a_b)
asja_basis = ['X' if x==1 else "Y" if x==0 else "Z" if x==2 else x for x in asja_basis]
print("Asja Basis:      ",asja_basis)


# Creating randomly chosen basis for Balvis (X,Y,Z)
for i in range(24):
    a_b = randrange(3)
    balvis_basis.append(a_b)
balvis_basis = ['X' if x==1 else "Y" if x==0 else "Z" if x==2 else x for x in balvis_basis]
print("Balvis Basis:        ",balvis_basis)


#Creating random bit string
for i in range(24):
    bit = randrange(2)
    send.append(bit)
print("The bits Asja wants to send:     ",send)

    
```

**Asja preparing states for sending**

```python
#Apply X gate if bit is equal to 1
for i, n in enumerate(send):
    if n==1:
        asja.x(qreg[i]) # apply x-gate
for i, k in enumerate(asja_basis): 
    if k=='X':
        asja.h(qreg[i])
    if k=="Y":
        asja.h(qreg[i])
        asja.s([qreg[i]])
            
        send_str = ''.join(str(e) for e in send)



asja.draw("mpl") #To check gates applied correctly
```

**Balvis measuring the sent states in random basis ($X$, Y or $Z$)**

```python
balvis = QuantumCircuit(qreg, creg, name='Balvis') #Defining Balvis circuit
SendState(asja, balvis, 'Asja') #Asja sends states to Balvis

for i, k in enumerate(balvis_basis): 
    if k=='X':
        balvis.h(qreg[i])
    if k=="Y":
        balvis.s(qreg[i])
        balvis.h(qreg[i])
        balvis.x(qreg[i])


            

balvis.draw("mpl")
```

**Backend measurnig of qubits in computational basis**

```python
balvis.measure(qreg,creg)

job = assemble(balvis)
result = Aer.get_backend('qasm_simulator').run(job).result()
counts = result.get_counts(balvis)
received = print_outcomes_in_reserve(counts)

print("Asja sent:", send_str)
print("Balvis received:", received)
```

**The sifted key Balvis and Asja would keep**

```python

asja_key=[] #Asjas register for matching rounds
balvis_key=[] #Balvis register for matching rounds
for j in range(0,len(asja_basis)): #Going through list of bases 
    if asja_basis[j] == balvis_basis[j]: #Comparing
        asja_key.append(send[j])
        balvis_key.append(received[j]) #Keeping key bit if bases matched
    else:
        pass #Discard round if bases mismatched
balvis_key = [eval(i) for i in balvis_key]

print("Asjas key =", asja_key)
print("Balvis key =", balvis_key)
        
```

In this protocol there was no evesdropping so both the Asja and Balvis keys are identical.
